# taking out a single row from a DataFrame

```python
import pandas as pd

import numpy as np
v = np.random.randint(56,90,16)
v1 = np.random.randint(60,90,16)
v2 = np.random.randint(60,90,16)
data = pd.DataFrame({"name":['bi.bi.kutheja','chandrushekar','darshini'
                             ,'deepak','harsha','manasa','manoj',
                             'monstar','pavan','nisarga','priya',
                             'prajwal','shakthi','spoorthi','yashwanth.R',
                             'yashwanth.K.G'],
                     'year1':v,
                     'year2':v1,
                     'year3':v2})

print(data)
print("Value of row 1")
print(data.iloc[7])
```

# aggreagation sum() min() max()

```python
print(data[data.year1==data.year1.max()])
print(data[data.year1==data.year1.min()])
print(data.year2)
```

# Group by

```python
data = pd.read_csv('diamonds.csv')
data
```

```python
data.head(7)
data.groupby(['cut']).mean()
```

```python
data.groupby(['cut','color']).agg('min')
```

```python
f = {'price':['sum','min','max','mean','median','prod']}
data.groupby(['color']).agg(f)
print(data.groupby(['color']).agg(f))
```
